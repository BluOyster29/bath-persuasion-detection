```python
import os 

folder_path = '/Users/rt853/repos/UoB/bath-persuasion-detection/data/phase_1_data/main_markup'
all_da_dfs = []
for batch_path in os.listdir(folder_path):

    batch = folder_path + '/' + batch_path
    print(batch)
    if not os.path.isdir(batch):
        continue
    
    for file in os.listdir(batch):
        
        file_name, file_type = file.split('.') 

        if file_type != 'jsonl':
            continue
        
        jsonl_path = batch + '/' + file
        jsonl = open_jsonl(jsonl_path)
        df = generate_dataframe(jsonl, label_columns, 'data')
        all_da_dfs.append(df)
```
