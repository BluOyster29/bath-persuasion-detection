```python
import pandas as pd 
```

```python
df = pd.read_csv('../../data/phase_2_data/processed/phase_2_partial_complete.csv')
```

```python
df
def generate_binary_datasets(df, label):
    for idx, row in df.iterrows():
        if row.gold == label:
            df.at[idx, 'binary_label'] = int(1)
        else:
            df.at[idx, 'binary_label'] = int(0)
    
    return df
```

```python
for label in df.gold.unique():
    bin_d = generate_binary_datasets(df, label)
    bin_d.to_csv(f'/Users/rt853/repos/UoB/bath-persuasion-detection/data/phase_2_data/processed/binary_datasets/{label}_binary_ds.csv', index=False)
```
