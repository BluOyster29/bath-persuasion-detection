```python
import sys 
import pandas as pd 

sys.path.append('../')
```

```python
def load_dialogues(path):
    with open(path) as file:
        lines = file.read()
        dialogues = lines.split('\n\n')
    return dialogues

def process_dialogues_frm_cgpt(dialogues):

    processed = []
    for d in dialogues:
        processed += [i for i in d.split('\n') if i != ''][2:]
    return processed
```

```python
dialogue_lines = load_dialogues('../data/testing/chat_gpt_prompt.txt')
```

```python
dialogue_lines = process_dialogues_frm_cgpt(dialogue_lines)
```

```python
def lines_to_df(dialogue_lines):
    data = {
        'text' : [],
        'label' : []
    }
    
    for i in dialogue_lines:
        label, text = i.split(': ')
        data['text'].append(text)
        data['label'].append(label)

    df = pd.DataFrame(data)
    for label in sorted(set(data['label'])):
        df[label] = pd.get_dummies(df['label'])[label]

    return df
```

```python
df = lines_to_df(dialogue_lines)
```

```python
from sklearn.preprocessing import LabelBinarizer

def convert_labels_to_one_hot(labels):
    lb = LabelBinarizer()
    one_hot_labels = lb.fit_transform(labels)
    return one_hot_labels
```
